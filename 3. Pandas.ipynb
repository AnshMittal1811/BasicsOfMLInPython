```python
d3 = pd.DataFrame({
    "city" : ["lucknow", "kanpur","Delhi","Agra"],
    "temperature" : [32,45,40,30]
})
```

```python
d3
```

```python
d4 = pd.DataFrame({
    "city" : ["Delhi","lucknow", "kanpur"],
    "Humidity" : [68,65,75]
})
```

```python
d4
```

```python
df6 = pd.merge(d3,d4, on = "city")
```

```python
df6
```

```python
df6 = pd.merge(d3,d4, on = "city",how = 'outer')
```

```python
df6
```

```python
df7 = pd.merge(d3,d4, on = ["city"],how = 'left')
```

```python
df7
```
